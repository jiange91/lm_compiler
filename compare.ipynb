```python
import matplotlib.pyplot as plt

# Data and settings
methods = ["Original", "DSPy", "Cognify", "Cognify", "Cognify"]
scores = [0.34149, 0.361, 0.381, 0.368, 0.375]
costs = [0.29123, 0.719, 0.5331927, 0.3441711, 0.4242966]

effs = [1 / x for x in costs]
# normalize efficiency
effs = [x / max(effs) for x in effs]
marker_size = 300

unique_methods = {"Original": "blue", "DSPy": "orange", "Cognify": "green"}
marker_styles = {"Original": "s", "DSPy": "o", "Cognify": "*"}


# Create scatter plot
plt.figure(figsize=(6, 4))
for method, score, eff in zip(methods, scores, effs):
    plt.scatter(
        eff,
        score,
        label=method if method not in plt.gca().get_legend_handles_labels()[1] else "", 
        color=unique_methods[method],
        marker=marker_styles[method],
        s=marker_size
    )
    
plt.annotate(
    '', xy=(effs[-3], scores[1]), xytext=(effs[-3], 0.378),
    arrowprops=dict(arrowstyle='<|-, head_width=0.5', color='red', lw=2)
)
score_impv = (scores[-3] - scores[1])
plt.text(0.45, (0.380 + scores[1]) / 2, f"+{score_impv:.2f}", fontsize=14, color="red")

plt.annotate(
    '', xy=(effs[1], scores[-2]), xytext=(0.8, scores[-2]),
    arrowprops=dict(arrowstyle='<|-, head_width=0.5', color='black', lw=2)
)
eff_impv = effs[-2] / effs[1]
plt.text((effs[1] + 0.8)/2, 0.365, f"{eff_impv:.1f}x", fontsize=14, color="black")


# Adding plot labels and legend
plt.ylabel("Quality", fontsize=14)
plt.xlabel("Cost-Effectiveness", fontsize=14)
plt.title("Hover", fontsize=16)
plt.legend(fontsize=14, loc="lower center", markerscale=0.7)
# plt.grid(True)
# plt.ylim(0, 0.3)
# plt.yticks([i * 0.05 for i in range(7)], fontsize=12)
# plt.xlim(0.33, 0.6)
# plt.xticks(fontsize=12)

# Show plot
plt.show()
```
