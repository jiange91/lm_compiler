```python
import matplotlib.pyplot as plt
```

```python
# Increase marker size for better visibility
methods = ["Original", "DSPy", "Cognify", "Cognify", "Cognify"]
quality = [1 - x for x in [0.56, 0.49, 0.44, 0.38, 0.35]]
costs = [0.06, 0.124, 0.07, 0.12, 0.26]
effs = [1 / x for x in costs]
# normalize efficiency
effs = [x / max(effs) for x in effs]

# Set a larger marker size
marker_size = 250

unique_methods = {"Original": "blue", "DSPy": "orange", "Cognify": "green"}
marker_styles = {"Original": "s", "DSPy": "o", "Cognify": "*"}  # Square, Circle, Star


# Create the scatter plot with larger markers for each method
plt.figure(figsize=(6, 4))
for method, score, eff in zip(methods, quality, effs):
    plt.scatter(
        eff,
        score,
        label=method if method not in plt.gca().get_legend_handles_labels()[1] else "", 
        color=unique_methods[method],
        marker=marker_styles[method],
        s=marker_size
    )

# Adding plot labels and legend
plt.ylabel("Quality", fontsize=14)
plt.xlabel("Cost-Effectiveness", fontsize=14)
plt.title("HotpotQA", fontsize=16)
plt.legend(fontsize=14, loc="upper right", markerscale=0.7)
# plt.grid(True)
# plt.ylim(0, 0.3)
# plt.yticks([i * 0.05 for i in range(7)], fontsize=12)
plt.yticks(fontsize=12)
# plt.xlim(0.33, 0.6)
plt.xticks(fontsize=12)

plt.annotate(
    '', xy=(effs[-1], quality[-1]-0.015), xytext=(effs[-1], quality[1]),
    arrowprops=dict(arrowstyle='-|>, head_width=0.75', color='red', lw=4),
)
# f1_improvement = ((quality[-1] - quality[1]) / quality[-1]) * 100
f1_improvement = (quality[-1] - quality[1])
plt.text(0.3, (quality[-1]-0.015 + quality[1]) / 2, f'+{f1_improvement:.2f}', 
                 ha='center', va='bottom', fontsize=14, color='red')

plt.annotate(
    '', xy=(effs[1], quality[2]), xytext=(0.8, quality[2]),
    arrowprops=dict(arrowstyle='<|-, head_width=0.75', color='black', lw=4),
)
eff_improvement = effs[2] / effs[1]
plt.text((effs[1] + 0.8) / 2, 0.54, f'{eff_improvement:.1f}x', 
                 ha='center', va='bottom', fontsize=14, color='black')

plt.show()
```
